```python
import pandas as pd
import cv2
import os
from tqdm import tqdm
from matplotlib import pyplot as plt
```

```python
filename='submission.csv'
df=pd.read_csv(filename)
df.columns=['vid','frame_id','bb_left','bb_top', 'width', 'height','class','confidence']
df
```

# Extract frame

```python
outframe='frame/'
src='videos/'
if not os.path.isdir(outframe):
    os.mkdir(outframe)
for file in tqdm(sorted(os.listdir(src))):
    video_id=int(file[:3])    
    vidcap = cv2.VideoCapture(src+file)
    count = 1
    while True:
        success,image = vidcap.read()
        if not success:
            break
#         cv2.imshow('window_name', image) 
        cv2.imwrite(outframe+'%03.0f' % video_id + "_%03.0f.jpg" % count, image)     # save frame as JPEG file          
        count += 1
    print(f'video_id: {video_id} - frame: {count}')
```

# Draw bbox

```python
def draw_bbox(image,x,y,xt,yt,label):
    start_point=(x,y)
    end_point=(xt,yt)
    # yellow color in BGR
    color = (255,255,0)
    thickness = 5
    image = cv2.rectangle(image, start_point, end_point, color, thickness)
    cv2.putText(image, label, (x-30, y+10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)
    return image
```

```python
outframe='bbox/'
src='frame/'
if not os.path.isdir(outframe):
    os.mkdir(outframe)
for file in tqdm(sorted(os.listdir(src))):
    if os.path.exists(outframe+file): continue
    vid_id,frame=[int(i) for i in file[:-4].split('_')]
    image = cv2.imread(src+file)
    for index, row in tqdm(df[(df['video_id']==vid_id) & (df['frame']==frame)].iterrows()):
        image=draw_bbox(image,int(row['bb_left']),int(row['bb_top']),int(row['bb_left']+row['bb_width']),int(row['bb_top']+row['bb_height']),str(row['class']))
#     plt.imshow(image)
#     plt.show()
    cv2.imwrite(outframe+file,image)
    
```

# Yolo format

```python
#define train file
ratio=0.9
train_dict={}
for vid_id in range(1,100+1):
    threshold=round(max(df[df['video_id']==vid_id]['frame']*ratio))#to determine the number of frame
    train_dict[vid_id]=threshold
train_dict
```

```python
# images
import shutil
src='frame/'
out='yolov8_tracking/datasets/images/'
train='train/'
val='val/'
for file in tqdm(os.listdir(src)): #None sorted for ramdomness
    vid_id,frame=[int(i) for i in file[:-4].split('_')]
    if train_dict[vid_id]>0:
        dst=out+train
        train_dict[vid_id]-=1
    else:
        dst=out+val
    shutil.copy(src+file, dst)
```

```python
# labels
src='datasets/images/'
out='datasets/labels/'
train='train/'
val='val/'
for file in tqdm(os.listdir(src+train)): #None sorted for ramdomness
    vid_id,frame=[int(i) for i in file[:-4].split('_')]
    image = cv2.imread(src+train+file)
    h,w,_=image.shape
    for index, row in tqdm(df[(df['video_id']==vid_id) & (df['frame']==frame)].iterrows()):
        #  class x_center y_center width height
        x=(row['bb_left']+row['bb_width']/2)/w
        y=(row['bb_top']+row['bb_height']/2)/h
        data=f'''{row["class"]} {x} {y} {row['bb_width']/w} {row['bb_height']/h}\n'''
        with open(out+train+file[:-4]+'.txt','a') as f:
            f.write(data)

for file in tqdm(os.listdir(src+val)): #None sorted for ramdomness
    vid_id,frame=[int(i) for i in file[:-4].split('_')]
    image = cv2.imread(src+val+file)
    h,w,_=image.shape
    for index, row in tqdm(df[(df['video_id']==vid_id) & (df['frame']==frame)].iterrows()):
        #  class x_center y_center width height
        x=(row['bb_left']+row['bb_width']/2)/w
        y=(row['bb_top']+row['bb_height']/2)/h
        data=f'''{row["class"]} {x} {y} {row['bb_width']/w} {row['bb_height']/h}\n'''
        with open(out+val+file[:-4]+'.txt','a') as f:
            f.write(data)
            
    
    
```
